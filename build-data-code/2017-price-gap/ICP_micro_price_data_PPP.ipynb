```python
import pandas as pd
import numpy as np
from scipy.stats import gmean
```

```python
file = "ICP-Researcher-Data_Global_2011-2017.xls"

cntry_names = ['AUS','AUT','BEL','BRA','CAN','CHN','COL','DNK','FIN','FRA','DEU'
            ,'GRC','IND','IDN','IRL','ITA','JPN','KOR','MEX','NOR','POL','PRT'
            ,'RUS','ESP','SWE','CHE','THA','TUR','GBR','USA', "Item Code","Description","Year"]

just_names = ['AUS','AUT','BEL','BRA','CAN','CHN','COL','DNK','FIN','FRA','DEU'
            ,'GRC','IND','IDN','IRL','ITA','JPN','KOR','MEX','NOR','POL','PRT'
            ,'RUS','ESP','SWE','CHE','THA','TUR','GBR','USA']
```

```python
dfexr = pd.read_excel(file, sheet_name = "EXR", header = 4)
```

```python
dfexr = dfexr[["Country Code", 2017]]

dfexr=  dfexr.loc[dfexr["Country Code"].isin(just_names), :]

dfexr.reset_index(inplace=True, drop=True)

dfexr.head()
```

```python
rotated_df = dfexr.set_index('Country Code').transpose()

rotated_df.reset_index(inplace=True)

rotated_df.rename(columns={"index": "Year"}, inplace=True)

rotated_df["Description"] = "EXR"

rotated_df["Item Code"] = 000

rotated_df["Traded"] = 1.0
```

```python
rotated_df.head()
```

```python
df = pd.read_excel(file, sheet_name = "PPP", header = 3)
```

```python
df.rename(columns = {'Unnamed: 1': 'Item Code', "Unnamed: 2": "Description", "Unnamed: 4": "Year"}, inplace = True)

df.drop(columns = ['Unnamed: 0',"Country Code", "Unnamed: 3"], inplace = True)

df.drop([0], inplace = True)

df.head()

filtered_df = df.loc[:, df.columns.isin(cntry_names)].copy(deep=True)

filtered_df = filtered_df[filtered_df["Year"] == 2017]
```

```python
filtered_df.head()
```

```python
df_traded = pd.read_csv("2017_catagories_traded.csv", header = 0)
```

```python
df_traded.head()
```

```python
bhprices = pd.merge(filtered_df, df_traded[["BH Name", "BH Code", "Traded"]], how = "right", 
                       left_on = "Item Code", right_on = "BH Code")
```

```python
bhprices.drop(columns = ["BH Name", "BH Code"], inplace = True)

bhprices.head()
```

```python
bhprices = pd.concat([bhprices, rotated_df], axis=0, ignore_index=True)
```

```python
bhprices_traded = bhprices[bhprices["Traded"] == 1.0].copy(deep=True)

bhprices_traded.describe()
```

```python
numeric_bhprices = bhprices_traded.select_dtypes(include=[np.number])

row_to_divide_by = numeric_bhprices.iloc[-1]

result = numeric_bhprices.div(row_to_divide_by, axis=1)

non_numeric_bhprices = bhprices_traded.select_dtypes(exclude=[np.number])

final_result = pd.concat([result, non_numeric_bhprices], axis=1)

final_result.tail()
```

```python
final_result.drop(columns = ["Traded", "Item Code", "Description", "Year"], inplace = True)

final_result.drop(final_result.index[-1], inplace = True)
```

```python
outdf = final_result[just_names].transpose()
```
