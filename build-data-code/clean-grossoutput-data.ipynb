```python
import pandas as pd             # data package
import matplotlib.pyplot as plt # graphics 
import datetime as dt
import numpy as np
import os

import requests, io             # internet and input tools  
import zipfile as zf            # zip file tools 
import os  

#import weightedcalcs as wc
#import numpy as np

import pyarrow as pa
import pyarrow.parquet as pq
```

```python
# bring in the data

df = pd.read_csv("./UNIDO-data/data.csv")

df = df[["Year", "Country", "CountryCode", "Activity", "ValueUSD"]]

df = df.rename(columns={"ValueUSD": "Value"})

out_file = "./UNIDO-data/UNIDO-grossoutput-data.parquet"

pq.write_table(pa.Table.from_pandas(df), out_file)
```

```python
year = 2004

dfyear = df[df["Year"] == year].copy(deep = True)
```

```python
dfyear.sort_values(by = ["Value"], ascending = False, inplace = True)

dfyear.head(40)

dfyear.loc[dfyear["Country"] == "Belgium", "Value"] = dfyear.loc[dfyear["Country"] == "Belgium", "Value"].iloc[0] + dfyear.loc[dfyear["Country"] == "Netherlands", "Value"].iloc[0]
```

```python
dfyear.head(40)
```

```python
# List of country codes
cntry_names = ['AUS', 'AUT', 'BEL', 'BRA', 'CAN', 'CHN', 'COL', 'DNK', 'FIN', 'FRA', 'DEU',
               'GRC', 'IND', 'IDN', 'IRL', 'ITA', 'JPN', 'KOR', 'MEX', 'NOR', 'POL', 'PRT',
               'RUS', 'ESP', 'SWE', 'CHE', 'THA', 'TUR', 'GBR', 'USA']

# List of country names
country_names = ['Australia', 'Austria', 'Belgium', 'Brazil', 'Canada', 'China', 'Colombia', 'Denmark', 'Finland', 'France', 'Germany',
                 'Greece', 'India', 'Indonesia', 'Ireland', 'Italy', 'Japan', 'Republic of Korea', 'Mexico', 'Norway', 'Poland', 'Portugal',
                 'Russian Federation', 'Spain', 'Sweden', 'Switzerland', 'Thailand', 'Türkiye', 'United Kingdom', 'United States of America']

# Create a dictionary mapping country names to codes
country_code_map = dict(zip(country_names, cntry_names))

# Filter the dataframe based on the country names
filtered_df = dfyear[dfyear['Country'].isin(country_names)].copy()

# Add a new column with the ISO codes
filtered_df['ISO_Code'] = filtered_df['Country'].map(country_code_map)

# # Save the filtered dataframe to a new file
# filtered_out_file = "./UNIDO-data/filtered-grossoutput-data.parquet"

# pq.write_table(pa.Table.from_pandas(filtered_df), filtered_out_file)

# # Display the first few rows of the filtered dataframe
filtered_df.head(30)
```

```python
# Ensure the ISO_Code column is sorted based on the order in cntry_names
filtered_df['ISO_Code'] = pd.Categorical(filtered_df['ISO_Code'], categories=cntry_names, ordered=True)

# Sort the DataFrame by the ISO_Code column
sorted_df = filtered_df.sort_values(by='ISO_Code')
```

```python
namesdf = pd.read_csv("top30-codes-names.csv")
```

```python
sorted_df = sorted_df.merge(namesdf, how = "left", left_on = "ISO_Code", right_on = "Country code")
```

```python
sorted_df.drop(columns = ["Country code"], inplace = True)
```

```python
# Display the sorted DataFrame
sorted_df.head(30)

sorted_df.reset_index(drop=True, inplace=True)

sorted_df.to_csv("./UNIDO-data/UNIDO-grossoutput-data-" + str(year) + ".csv", index=False)
```
